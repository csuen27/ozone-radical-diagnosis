```python
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['Arial']
```

```python
budget_season = pd.read_csv('Fig. 1 radical budget by season.csv')
```

```python
budget_season
```

### O(1D) radical budget

```python
f = plt.figure(num=None, figsize=(2, 10), dpi=100, facecolor='w', edgecolor='k')

S = budget_season['Time']
prod1 = budget_season['P1']*10**(-6)
loss1 = -budget_season['P2']*10**(-6)
loss2 = -budget_season['P3']*10**(-6)
loss3 = -budget_season['P4']*10**(-6)

a = 0.75

plt.axhline(y=0, color='black', linestyle='--', linewidth=0.8)
plt.bar(S, prod1, width=a, color='salmon', edgecolor='black')
plt.bar(S, loss1, width=a, color='mistyrose', bottom=0, edgecolor='black')
plt.bar(S, loss2, width=a, color='white', bottom=loss1, edgecolor='black')
plt.bar(S, loss3, width=a, color='silver', bottom=loss1+loss2, edgecolor='black')
plt.ylabel("O(${^1}$D) chemical budget compositions by season", fontsize=16)

plt.yticks(fontsize=16)
plt.xticks(fontsize=16, rotation=90)

ax=plt.gca()
#ax.spines['top'].set_linewidth(0)
#ax.spines['right'].set_linewidth(0)

plt.show()
#f.savefig("Fig. 1a decomposition of O(1D) budget.pdf", bbox_inches="tight")
```

```python
### OH radical budget
```

```python
f = plt.figure(num=None, figsize=(2, 10), dpi=100, facecolor='w', edgecolor='k')

S = budget_season['Time']

prod1 = budget_season['P5']*10**(-6)
prod2 = budget_season['P6']*10**(-6)
prod3 = budget_season['P7']*10**(-6)

loss1 = -budget_season['P8']*10**(-6)
loss2 = -budget_season['P9']*10**(-6)
loss3 = -budget_season['P10']*10**(-6)
loss4 = -budget_season['P11']*10**(-6)
loss5 = -budget_season['P12']*10**(-6)
loss6 = -budget_season['P13']*10**(-6)

a = 0.75

plt.axhline(y=0, color='black', linestyle='--', linewidth=0.8)
plt.bar(S, prod1, width=a, color='salmon', edgecolor='black')
plt.bar(S, prod2, width=a, color='lemonchiffon', bottom=prod1, edgecolor='black')
plt.bar(S, prod3, width=a, color='paleturquoise', bottom=prod1+prod2, edgecolor='black')

plt.bar(S, loss1, width=a, color='mistyrose', bottom=0, edgecolor='black')
plt.bar(S, loss2, width=a, color='white', bottom=loss1, edgecolor='black')
plt.bar(S, loss3, width=a, color='silver', bottom=loss1+loss2, edgecolor='black')
plt.bar(S, loss4, width=a, color='mistyrose', bottom=loss1+loss2+loss3, edgecolor='black')
plt.bar(S, loss5, width=a, color='white', bottom=loss1+loss2+loss3+loss4, edgecolor='black')
plt.bar(S, loss6, width=a, color='silver', bottom=loss1+loss2+loss3+loss4+loss5, edgecolor='black')
plt.ylabel("OH chemical budget compositions by season", fontsize=16)

plt.yticks(fontsize=16)
plt.xticks(fontsize=16, rotation=90)

ax=plt.gca()
#ax.spines['top'].set_linewidth(0)
#ax.spines['right'].set_linewidth(0)

plt.show()
#f.savefig("Fig. 1b decomposition of OH budget.pdf", bbox_inches="tight")
```
