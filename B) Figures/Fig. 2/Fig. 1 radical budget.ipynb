```python
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['Arial']
```

```python
budget_season = pd.read_csv('Fig. 1 radical budget by season.csv')
```

```python
data = budget_season[budget_season['Time']=="Annual"]
```

### O(${^1}$D) radical budget

```python
f = plt.figure(num=None, figsize=(2, 10), dpi=100, facecolor='w', edgecolor='k')

S = budget_season['Time']
prod1 = budget_season['P1']*10**(-6)
loss1 = -budget_season['P2']*10**(-6)
loss2 = -budget_season['P3']*10**(-6)
loss3 = -budget_season['P4']*10**(-6)

a = 0.75

plt.axhline(y=0, color='black', linestyle='--', linewidth=0.8)
plt.bar(S, prod1, width=a, color='salmon', edgecolor='black')
plt.bar(S, loss1, width=a, color='mistyrose', bottom=0, edgecolor='black')
plt.bar(S, loss2, width=a, color='white', bottom=loss1, edgecolor='black')
plt.bar(S, loss3, width=a, color='silver', bottom=loss1+loss2, edgecolor='black')
plt.ylabel("O(${^1}$D) chemical budget compositions by season", fontsize=16)

plt.yticks(fontsize=16)
plt.xticks(fontsize=16, rotation=90)

ax=plt.gca()
#ax.spines['top'].set_linewidth(0)
#ax.spines['right'].set_linewidth(0)

plt.show()
#f.savefig("Fig. 1a decomposition of O(1D) budget.pdf", bbox_inches="tight")
```
