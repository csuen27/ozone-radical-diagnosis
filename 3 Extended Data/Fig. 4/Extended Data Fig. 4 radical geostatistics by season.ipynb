```python
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('D:/Research/radicals/UKESM1-0-LL_2010_2014.csv')
land = pd.read_csv("C:/Users/Haitong/OneDrive - National University of Singapore/Documents/Manuscripts/Sun-NG-25/global_grid_2x2_label_continent.csv")
dataset_land = dataset.merge(land, how='left', left_on=['longitude','latitude'], right_on=['longitude','latitude'])
dataset_land = dataset_land[dataset_land['land_global']==1]
```

### Geomapping the radical concentration

```python
temp = dataset.groupby('loc').mean()

plt.figure(num=None, figsize=(10, 6), dpi=50, facecolor='w', edgecolor='k')
cm = plt.cm.get_cmap('viridis')
x = temp['longitude']
y = temp['latitude']

plt.subplot(2,2,1)
plt.title('O(${^1}$D)', fontsize = 12, fontname = "Arial", color = "black")
z = (temp['o1d_mole_adj'])*10**3
sc = plt.scatter(x, y, c=z, marker="s", s=5, cmap=cm)
plt.xlim([-180, 180])
plt.ylim([-90, 90])
plt.colorbar(fraction=0.02)

plt.subplot(2,2,2)
plt.title('OH', fontsize = 12, fontname = "Arial", color = "black")
z = (temp['oh_mole'])*10**(-6)
sc = plt.scatter(x, y, c=z, marker="s", s=5, cmap=cm)
plt.xlim([-180, 180])
plt.ylim([-90, 90])
plt.colorbar(fraction=0.02)

plt.subplot(2,2,3)
plt.title('HO${_2}$', fontsize = 12, fontname = "Arial", color = "black")
z = (temp['ho2_mole_adj'])*10**(-8)
sc = plt.scatter(x, y, c=z, marker="s", s=5, cmap=cm)
plt.xlim([-180, 180])
plt.ylim([-90, 90])
plt.colorbar(fraction=0.02)

plt.subplot(2,2,4)
plt.title('CH${_3}$O${_2}$', fontsize = 12, fontname = "Arial", color = "black")
z = (temp['ch3o2_mole_adj'])*10**(-8)
sc = plt.scatter(x, y, c=z, marker="s", s=5, cmap=cm)
plt.xlim([-180, 180])
plt.ylim([-90, 90])
plt.colorbar(fraction=0.02)
```
