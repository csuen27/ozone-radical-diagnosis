```python
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['Arial']
```

```python
o3prod = pd.read_csv('Fig. 2 o3prod.csv')
```

```python
o3prod_colors = ['darkslategrey','palegoldenrod','turquoise']

R=1.5
r1=0.4
```

```python
fig, ax = plt.subplots(figsize=(10, 6), dpi=75)
labels = ['HO${_2}$+NO','CH${_3}$O${_2}$+NO','RO${_2}$+NO']
budgets_JJA = o3prod['NA_JJA'].values
centre=[1]

ax.pie(budgets_JJA, radius=R, labels=labels, colors=o3prod_colors, startangle=90)
ax.pie(centre, radius=R-r1,colors='w')

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6), dpi=75)
budgets_JJA = o3prod['NA_JJA'].values
centre=[1]

ax.pie(budgets_JJA, radius=R, labels=None, colors=o3prod_colors, startangle=90)
ax.pie(centre, radius=R-r1,colors='w')

fig.savefig("Extended Data Fig. 6a-1 Production Rate North America (JJA).pdf", bbox_inches="tight")
```
