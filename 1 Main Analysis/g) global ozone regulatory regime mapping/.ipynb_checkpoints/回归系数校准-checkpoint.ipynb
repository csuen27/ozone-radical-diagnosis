```python
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['Arial']
import statsmodels.api as sm
import statsmodels.formula.api as smf
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
```

```python
data = pd.read_csv('regime diagnosis map.csv')
data
```

```python
plt.scatter(np.log(data['LER_etoo']), np.log(data['FNR_NO2']))
```

```python
data['ln_FNR_NO2'] = np.log(data['FNR_NO2'])
data['ln_FNR_NOx'] = np.log(data['FNR_NOx'])
data['ln_FNR_NOy'] = np.log(data['FNR_NOy'])
data['ln_HOP'] = np.log(data['HOP'])
data['ln_LER'] = np.log(data['LER_etoo'])
```

```python
mod = LinearRegression()
y = data['SnSv_adj2']
X = data[['ln_HOP','ln_LER','ln_FNR_NO2','ln_FNR_NOy','ln_FNR_NOx']]
mod.fit(X,y)
y_pred = mod.predict(X)
y_pred
```

```python
plt.figure(num=None, figsize=(20, 2.75), dpi=100, facecolor='w', edgecolor='k')
cm = plt.cm.get_cmap('Reds')
x = data['longitude']
y = data['latitude']

plt.subplot(1,3,1)
plt.title('SnSv', fontsize = 20, fontname = "Arial", color = "black")
z = data['Sv_adj']
sc = plt.scatter(x, y, c=z, marker="s", s=2, cmap=cm)
plt.xlim([-180, 180])
plt.ylim([-60, 75])
plt.colorbar(fraction=0.02)
```

```python
country_mean['regime_FNR']=0
country_mean['regime_HOP']=0
country_mean['regime_LER']=0
country_mean['regime_REG']=0

for t in range(0,153,1):
    
    SnSv_Reg = country_mean.iloc[t,7]
    SnSv_FNR = country_mean.iloc[t,20]
    SnSv_HOP = country_mean.iloc[t,21]
    SnSv_LER = country_mean.iloc[t,22]
    
    if SnSv_FNR>=1.2:
        regime = 4
    elif 0.8<=SnSv_FNR<1.2:
        regime = 3
    elif 0<=SnSv_FNR<0.8:
        regime = 2
    else: 
        regime = 1
        
    country_mean.iloc[t,23] = regime
    
    if SnSv_HOP>=1.2:
        regime = 4
    elif 0.8<=SnSv_HOP<1.2:
        regime = 3
    elif 0<=SnSv_HOP<0.8:
        regime = 2
    else: 
        regime = 1
        
    country_mean.iloc[t,24] = regime
    
    if SnSv_LER>=1.2:
        regime = 4
    elif 0.8<=SnSv_LER<1.2:
        regime = 3
    elif 0<=SnSv_LER<0.8:
        regime = 2
    else: 
        regime = 1
        
    country_mean.iloc[t,25] = regime
    
    if SnSv_Reg>=1.2:
        regime = 4
    elif 0.8<=SnSv_Reg<1.2:
        regime = 3
    elif 0<=SnSv_Reg<0.8:
        regime = 2
    else: 
        regime = 1
        
    country_mean.iloc[t,26] = regime
```
