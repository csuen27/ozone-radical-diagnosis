```python
o3prod = pd.read_csv('/Users/csuen/Documents/GitHub/ozone-radical/Fig. 2/Fig. 2 o3prod.csv')
o3loss = pd.read_csv('/Users/csuen/Documents/GitHub/ozone-radical/Fig. 2/Fig. 2 o3loss.csv')
```
