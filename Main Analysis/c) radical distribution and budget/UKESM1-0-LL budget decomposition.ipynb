```python
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

dataset_land = pd.read_csv('/Volumes/Database/Research/ozone-budget/UKESM1-0-LL_2010_2014_land(ver231213).csv')
```

```python
def Pearson_correlation(X,Y):
	if len(X)==len(Y):
		Sum_xy = sum((X-X.mean())*(Y-Y.mean()))
		Sum_x_squared = sum((X-X.mean())**2)
		Sum_y_squared = sum((Y-Y.mean())**2)	 
		corr = Sum_xy / np.sqrt(Sum_x_squared * Sum_y_squared)
	return corr
```

```python
Pearson_correlation(dataset_land['o3prod_mole_final'], dataset_land['o3loss_mole_final'])
```
