```python
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['Arial']
```

```python
budget_region = pd.read_csv('Fig. 1 radical budget by region.csv')
```

```python
budget_region
```

### O(${^1}$D) radical budget

```python
f = plt.figure(num=None, figsize=(3, 10), dpi=100, facecolor='w', edgecolor='k')

S = budget_region['Region']
prod1 = budget_region['P1']*10**(-6)
loss1 = -budget_region['P2']*10**(-6)
loss2 = -budget_region['P3']*10**(-6)
loss3 = -budget_region['P4']*10**(-6)

a = 0.75

plt.axhline(y=0, color='black', linestyle='-', linewidth=2)

plt.bar(S, prod1, width=a, color='midnightblue', edgecolor='white', lw=2)
plt.bar(S, loss1, width=a, color='lightsteelblue', bottom=0, edgecolor='white', lw=2)
plt.bar(S, loss2, width=a, color='paleturquoise', bottom=loss1, edgecolor='white', lw=2)
plt.bar(S, loss3, width=a, color='mediumturquoise', bottom=loss1+loss2, edgecolor='white', lw=2)

ax=plt.gca()

plt.yticks([-12, -9, -6, -3, 0, 3, 6, 9, 12], fontsize=16)
plt.xticks(fontsize=16, rotation=90)
plt.ylim(-12, 12)

f.savefig("Extended Data Fig. 5a decomposition of O(1D) budget by region.pdf", bbox_inches="tight")
```

```python
budget_region['P1p'] = budget_region['P1']/budget_region['P1']
budget_region['P2p'] = budget_region['P2']/(budget_region['P2']+budget_region['P3']+budget_region['P4'])
budget_region['P3p'] = budget_region['P3']/(budget_region['P2']+budget_region['P3']+budget_region['P4'])
budget_region['P4p'] = budget_region['P4']/(budget_region['P2']+budget_region['P3']+budget_region['P4'])

f = plt.figure(num=None, figsize=(3, 10), dpi=100, facecolor='w', edgecolor='k')

S = budget_region['Region']
prod1 =  budget_region['P1p']
loss1 = -budget_region['P2p']
loss2 = -budget_region['P3p']
loss3 = -budget_region['P4p']

a = 0.75

plt.axhline(y=0, color='black', linestyle='-', linewidth=2)

plt.bar(S, prod1, width=a, color='midnightblue', edgecolor='white', lw=2)
plt.bar(S, loss1, width=a, color='lightsteelblue', bottom=0, edgecolor='white', lw=2)
plt.bar(S, loss2, width=a, color="paleturquoise", bottom=loss1, edgecolor='white', lw=2)
plt.bar(S, loss3, width=a, color='mediumturquoise', bottom=loss1+loss2, edgecolor='white', lw=2)

ax=plt.gca()

plt.yticks(fontsize=16)
plt.xticks(fontsize=16, rotation=90)
plt.ylim(-1,1)

f.savefig("Extended Data Fig. 5e decomposition proportion of O(1D) budget by region.pdf", bbox_inches="tight")
```

### OH radical budget

```python
f = plt.figure(num=None, figsize=(3, 10), dpi=100, facecolor='w', edgecolor='k')

S = budget_region['Region']

prod1 = budget_region['P6']*10**(-6)
prod2 = budget_region['P5']*10**(-6)
prod3 = budget_region['P7']*10**(-6)

loss1 = -budget_region['P8']*10**(-6)
loss2 = -budget_region['P9']*10**(-6)
loss3 = -budget_region['P10']*10**(-6)
loss4 = -budget_region['P11']*10**(-6)
loss5 = -budget_region['P12']*10**(-6)
loss6 = -budget_region['P13']*10**(-6)
loss456 = loss4 + loss5 + loss6

a = 0.75

plt.axhline(y=0, color='black', linestyle='-', linewidth=2)
plt.bar(S, prod1, width=a, color='midnightblue', edgecolor='white', lw=2)
plt.bar(S, prod2, width=a, color='thistle', bottom=prod1, edgecolor='white', lw=2)
plt.bar(S, prod3, width=a, color='gainsboro', bottom=prod1+prod2, edgecolor='white', lw=2)
plt.bar(S, loss1, width=a, color='lightsteelblue', bottom=0, edgecolor='white', lw=2)
plt.bar(S, loss2, width=a, color="paleturquoise", bottom=loss1, edgecolor='white', lw=2)
plt.bar(S, loss3, width=a, color='mediumturquoise', bottom=loss1+loss2, edgecolor='white', lw=2)
plt.bar(S, loss456, width=a, color='slategrey', bottom=loss1+loss2+loss3, edgecolor='white', lw=2)

ax=plt.gca()

plt.yticks([-4.8, -3.6, -2.4, -1.2, 0, 1.2, 2.4, 3.6, 4.8], fontsize=16)
plt.xticks(fontsize=16, rotation=90)
plt.ylim(-4.8, 4.8)

f.savefig("Extended Data Fig. 5b decomposition of OH budget by region.pdf", bbox_inches="tight")
```

```python
budget_region['P5p'] = budget_region['P5']/(budget_region['P5']+budget_region['P6']+budget_region['P7'])
budget_region['P6p'] = budget_region['P6']/(budget_region['P5']+budget_region['P6']+budget_region['P7'])
budget_region['P7p'] = budget_region['P7']/(budget_region['P5']+budget_region['P6']+budget_region['P7'])

budget_region['P8p'] = budget_region['P8']/(budget_region['P8']+budget_region['P9']+budget_region['P10']+budget_region['P11']+budget_region['P12']+budget_region['P13'])
budget_region['P9p'] = budget_region['P9']/(budget_region['P8']+budget_region['P9']+budget_region['P10']+budget_region['P11']+budget_region['P12']+budget_region['P13'])
budget_region['P10p'] = budget_region['P10']/(budget_region['P8']+budget_region['P9']+budget_region['P10']+budget_region['P11']+budget_region['P12']+budget_region['P13'])
budget_region['P11p'] = budget_region['P11']/(budget_region['P8']+budget_region['P9']+budget_region['P10']+budget_region['P11']+budget_region['P12']+budget_region['P13'])
budget_region['P12p'] = budget_region['P12']/(budget_region['P8']+budget_region['P9']+budget_region['P10']+budget_region['P11']+budget_region['P12']+budget_region['P13'])
budget_region['P13p'] = budget_region['P13']/(budget_region['P8']+budget_region['P9']+budget_region['P10']+budget_region['P11']+budget_region['P12']+budget_region['P13'])

f = plt.figure(num=None, figsize=(3, 10), dpi=100, facecolor='w', edgecolor='k')

S = budget_region['Region']

prod1 = budget_region['P6p']
prod2 = budget_region['P5p']
prod3 = budget_region['P7p']

loss1 = -budget_region['P8p']
loss2 = -budget_region['P9p']
loss3 = -budget_region['P10p']
loss4 = -budget_region['P11p']
loss5 = -budget_region['P12p']
loss6 = -budget_region['P13p']
loss456 = loss4 + loss5 + loss6

a = 0.75

plt.axhline(y=0, color='black', linestyle='-', linewidth=2)
plt.bar(S, prod1, width=a, color='midnightblue', edgecolor='white', lw=2)
plt.bar(S, prod2, width=a, color='thistle', bottom=prod1, edgecolor='white', lw=2)
plt.bar(S, prod3, width=a, color='gainsboro', bottom=prod1+prod2, edgecolor='white', lw=2)
plt.bar(S, loss1, width=a, color='lightsteelblue', bottom=0, edgecolor='white', lw=2)
plt.bar(S, loss2, width=a, color="paleturquoise", bottom=loss1, edgecolor='white', lw=2)
plt.bar(S, loss3, width=a, color='mediumturquoise', bottom=loss1+loss2, edgecolor='white', lw=2)
plt.bar(S, loss456, width=a, color='slategrey', bottom=loss1+loss2+loss3, edgecolor='white', lw=2)

ax=plt.gca()

plt.yticks(fontsize=16)
plt.xticks(fontsize=16, rotation=90)
plt.ylim(-1,1)

f.savefig("Extended Data Fig. 5f decomposition proportion of OH budget by region.pdf", bbox_inches="tight")
```

### HO${_2}$ radical budget

```python
f = plt.figure(num=None, figsize=(3, 10), dpi=100, facecolor='w', edgecolor='k')

S = budget_region['Region']

prod1 = budget_region['P14']*10**(-6)
prod2 = budget_region['P15']*10**(-6)
prod3 = budget_region['P16']*10**(-6)
prod4 = budget_region['P17']*10**(-6)
prod34 = prod3 + prod4

loss1 = -budget_region['P18']*10**(-6)
loss2 = -budget_region['P19']*10**(-6)
loss3 = -budget_region['P20']*10**(-6)
loss4 = -budget_region['P21']*10**(-6)
loss5 = -budget_region['P22']*10**(-6)
loss2345 = loss2 + loss3 + loss4 + loss5

a = 0.75

plt.axhline(y=0, color='black', linestyle='-', linewidth=2)

plt.bar(S, prod1, width=a, color='midnightblue', edgecolor='white', lw=2)
plt.bar(S, prod2, width=a, color='thistle', bottom=prod1, edgecolor='white', lw=2)
plt.bar(S, prod34, width=a, color='gainsboro', bottom=prod1+prod2, edgecolor='white', lw=2)
plt.bar(S, loss1, width=a, color='lightsteelblue', bottom=0, edgecolor='white', lw=2)
plt.bar(S, loss2345, width=a, color="mediumturquoise", bottom=loss1, edgecolor='white', lw=2)

ax=plt.gca()

plt.yticks([-2.4, -1.8, -1.2, -0.6, 0, 0.6, 1.2, 1.8, 2.4], fontsize=16)
plt.xticks(fontsize=16, rotation=90)
plt.ylim(-2.4,2.4)

f.savefig("Extended Data Fig. 5c decomposition of HO2 budget by region.pdf", bbox_inches="tight")
```

```python
budget_region['P14p'] = budget_region['P14']/(budget_region['P14']+budget_region['P15']+budget_region['P16']+budget_region['P17'])
budget_region['P15p'] = budget_region['P15']/(budget_region['P14']+budget_region['P15']+budget_region['P16']+budget_region['P17'])
budget_region['P16p'] = budget_region['P16']/(budget_region['P14']+budget_region['P15']+budget_region['P16']+budget_region['P17'])
budget_region['P17p'] = budget_region['P17']/(budget_region['P14']+budget_region['P15']+budget_region['P16']+budget_region['P17'])

budget_region['P18p'] = budget_region['P18']/(budget_region['P18']+budget_region['P19']+budget_region['P20']+budget_region['P21']+budget_region['P22'])
budget_region['P19p'] = budget_region['P19']/(budget_region['P18']+budget_region['P19']+budget_region['P20']+budget_region['P21']+budget_region['P22'])
budget_region['P20p'] = budget_region['P20']/(budget_region['P18']+budget_region['P19']+budget_region['P20']+budget_region['P21']+budget_region['P22'])
budget_region['P21p'] = budget_region['P21']/(budget_region['P18']+budget_region['P19']+budget_region['P20']+budget_region['P21']+budget_region['P22'])
budget_region['P22p'] = budget_region['P22']/(budget_region['P18']+budget_region['P19']+budget_region['P20']+budget_region['P21']+budget_region['P22'])

f = plt.figure(num=None, figsize=(3, 10), dpi=100, facecolor='w', edgecolor='k')

S = budget_region['Region']

prod1 = budget_region['P14p']
prod2 = budget_region['P15p']
prod3 = budget_region['P16p']
prod4 = budget_region['P17p']
prod34 = prod3 + prod4

loss1 = -budget_region['P18p']
loss2 = -budget_region['P19p']
loss3 = -budget_region['P20p']
loss4 = -budget_region['P21p']
loss5 = -budget_region['P22p']
loss2345 = loss2 + loss3 + loss4 + loss5

a = 0.75

plt.axhline(y=0, color='black', linestyle='-', linewidth=2)

plt.bar(S, prod1, width=a, color='midnightblue', edgecolor='white', lw=2)
plt.bar(S, prod2, width=a, color='thistle', bottom=prod1, edgecolor='white', lw=2)
plt.bar(S, prod34, width=a, color='gainsboro', bottom=prod1+prod2, edgecolor='white', lw=2)
plt.bar(S, loss1, width=a, color='lightsteelblue', bottom=0, edgecolor='white', lw=2)
plt.bar(S, loss2345, width=a, color="mediumturquoise", bottom=loss1, edgecolor='white', lw=2)

ax=plt.gca()

plt.yticks(fontsize=16)
plt.xticks(fontsize=16, rotation=90)
plt.ylim(-1,1)

f.savefig("Extended Data Fig. 5g decomposition proportion of HO2 budget by region.pdf", bbox_inches="tight")
```
