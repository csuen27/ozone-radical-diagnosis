```python
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['Arial']
```

```python
o3loss = pd.read_csv('Fig. 2 o3loss.csv')
```

```python
o3loss_colors = ['midnightblue','gold','silver','steelblue']

R=0.75
r1=0.2
```

```python
fig, ax = plt.subplots(figsize=(10, 6), dpi=75)
labels = ['O${^1}$D+H${_2}$O','O${_3}$+OH','O${_3}$+HO${_2}$','O${_3}$+VOCs']
budgets_JJA = o3loss['NA_JJA'].values
centre=[1]

ax.pie(budgets_JJA, radius=R, labels=labels, colors=o3loss_colors, startangle=90)
ax.pie(centre, radius=R-r1,colors='w')

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6), dpi=75)
budgets_JJA = o3loss['NA_JJA'].values
centre=[1]

ax.pie(budgets_JJA, radius=R, labels=None, colors=o3loss_colors, startangle=90)
ax.pie(centre, radius=R-r1,colors='w')

fig.savefig("Extended Data Fig. 7a-1 Loss Rate North America (JJA).pdf", bbox_inches="tight")
```
